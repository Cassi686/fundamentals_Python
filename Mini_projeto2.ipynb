```python
# Importando biblotecas
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Baixando dataframe
df=pd.read_csv('/content/Walmart (1).csv')
```

```python
# Visualizando dataframe
df
```

```python
# Procurando dados nulos
df.isnull().sum()
```

```python
# Analisando o dataframe
df.describe()
```

```python
# Localizando outliers do dataframe
sns.boxplot(df['Weekly_Sales']);
```

```python
# Agrupando dataframe por loja e obtendo preço média de vendas por semana.
df_groupby=df.groupby(['Store']).mean().reset_index()
df_groupby=df_groupby.drop(['Holiday_Flag','Temperature','Fuel_Price','CPI','Unemployment'],axis=1)
df_groupby
```

```python
#Obtendo soma de vendas por semana por loja.
df_groupby1=df.groupby(['Store']).sum().reset_index()
df_groupby1=df_groupby1.drop(['Holiday_Flag','Temperature','Fuel_Price','CPI','Unemployment'],axis=1)
df_groupby1
```

```python
#Obtendo a mediana de vendas por semana por loja.
df_groupby2=df.groupby(['Store']).median().reset_index()
df_groupby2
```

```python
# Obtendo tabela em ordem decrescente das 5 lojas com maior média de vendas por semana.
df_groupby.sort_values(by='Weekly_Sales', ascending=False).head(5)
```

```python
# Obtendo tabela em ordem decrescente das 5 lojas com maior soma de vendas por semana.
df_groupby1.sort_values(by='Weekly_Sales', ascending=False).head(5)
```

```python
#Obtendo tabela em ordem decrescente das 5 lojas com maiores medianas de vendas por semana.
df_groupby2.sort_values(by='Weekly_Sales', ascending=False).head(5)
```

```python
#Média geral do dataframe de vendas por semana
df['Weekly_Sales'].mean()
```

```python
#Quantidade de semanas em que a loja com maior vendas acumuladas ultrapassou a média do período.
select = df.loc[df['Store'] == 20]
select=select.loc[df['Weekly_Sales']>1046964.8775617715]
len(select)
```

```python
# Média de vendas por semana da loja com maior vendas acumuladas.
select = df.loc[df['Store'] == 20]
select_mean=select['Weekly_Sales'].mean()
select_mean
```

```python
# Quantidade de semanas que a loja com maior mediana de vendas por semana superou a sua própria média de vendas no período.
select = df.loc[df['Store'] == 4]
select=select.loc[df['Weekly_Sales']>2094712.9606993007]
len(select)
```
